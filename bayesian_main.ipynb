```python
import pandas as pd

dataset = pd.read_csv('16k_Movies.csv')
```

```python
dataset.head()
```

```python
dataset = dataset[['Title', 'Release Date', 'Rating']]

dataset
```

```python
# drop rows with missing values
dataset = dataset.dropna()
dataset
```

```python
# Keep only the year from the release date
dataset['Release Date'] = dataset['Release Date'].apply(lambda x: x[-4:])
dataset
```

```python
dataset['Title'] = dataset['Title'].astype(str)
dataset['Release Date'] = dataset['Release Date'].astype(int)
dataset
```

```python
dataset = dataset[dataset['Release Date'] >= 2018]
dataset
```

```python
# Categories: Rating >= 7.5 is good, Rating < 5 is bad,  5 <= Rating < 7.5 is average 
dataset['Review'] = dataset['Rating'].apply(lambda x: 'good' if x >= 7.5 else ('bad' if x < 5 else 'average'))
dataset['Recommendation'] = dataset['Rating'].apply(lambda x: 'Recommended' if x >= 6.5 else 'Not Recommended')
dataset
```

```python
# Save the dataset
dataset.to_csv('16k_Movies_cleaned.csv', index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Prior distribution: Flat prior
alpha_prior_1 = 1
beta_prior_1 = 1
X = np.linspace(0, 1, 1000)
Y = stats.beta.pdf(X, alpha_prior_1, beta_prior_1)
plt.plot(X, Y, label=f'Prior Distribution, Beta({alpha_prior_1}, {beta_prior_1})')

# Likelihood distribution
n = sum(dataset['Release Date'] == 2023)
y = sum(dataset[dataset['Release Date'] == 2023]['Recommendation'] == 'Recommended')
X = np.linspace(0, 1, 1000)
Y = stats.binom.pmf(y, n, X)
plt.plot(X, Y*200, label='Scaled Likelihood Distribution')

# Plot the posterior distribution
alpha_posterior_1 = alpha_prior_1 + y
beta_posterior_1 = beta_prior_1 + n - y
X = np.linspace(0, 1, 1000)
Y = stats.beta.pdf(X, alpha_posterior_1, beta_posterior_1)
plt.plot(X, Y, label=f'Posterior Distribution, Beta({alpha_posterior_1}, {beta_posterior_1})')

plt.xlabel('Probability of A Movie Being a Must Watch')
plt.ylabel('Density')
plt.title('Flat Prior')
lower_bound = stats.beta.ppf(0.025, alpha_posterior_1, beta_posterior_1)
upper_bound = stats.beta.ppf(0.975, alpha_posterior_1, beta_posterior_1)
print(f'95% Credible Interval: ({lower_bound}, {upper_bound})')
plt.axvline(lower_bound, color='red', linestyle='--', label='95% Credible Interval')
plt.axvline(upper_bound, color='red', linestyle='--')
plt.legend()
plt.show()

print(alpha_prior_1, beta_prior_1, alpha_posterior_1, beta_posterior_1)
print(n, y)
```

```python
year = 2018
historic_data = dataset[(dataset['Release Date'] >= year) & (dataset['Release Date'] <= 2022)]

# Prior distribution: Prior based on historic data
alpha_prior_2 = sum(historic_data['Recommendation'] == 'Recommended')
beta_prior_2 = len(historic_data) - alpha_prior_2
X = np.linspace(0, 1, 1000)
Y = stats.beta.pdf(X, alpha_prior_2, beta_prior_2)
plt.plot(X, Y, label=f'Prior, Beta({alpha_prior_2}, {beta_prior_2})')

# Likelihood distribution
n = sum(dataset['Release Date'] == 2023)
y = sum(dataset[dataset['Release Date'] == 2023]['Recommendation'] == 'Recommended')
X = np.linspace(0, 1, 1000)
Y = stats.binom.pmf(y, n, X)
plt.plot(X, Y*400, label='Scaled Likelihood')

# Plot the posterior distribution
alpha_posterior_2 = alpha_prior_2 + y
beta_posterior_2 = beta_prior_2 + n - y
X = np.linspace(0, 1, 1000)
Y = stats.beta.pdf(X, alpha_posterior_2, beta_posterior_2)
plt.plot(X, Y, label=f'Posterior, Beta({alpha_posterior_2}, {beta_posterior_2})')

# Find the 95% credible interval
lower_bound = stats.beta.ppf(0.025, alpha_posterior_2, beta_posterior_2)
upper_bound = stats.beta.ppf(0.975, alpha_posterior_2, beta_posterior_2)
print(f'95% Credible Interval: ({lower_bound}, {upper_bound})')
plt.axvline(lower_bound, color='red', linestyle='--', label='95% Credible Interval')
plt.axvline(upper_bound, color='red', linestyle='--')
plt.xlabel('Probability of A Movie Being a Must Watch')
plt.ylabel('Density')
plt.title(f'Prior Based on Historic Data ({year}-2022)')
plt.xlim(0.2, 0.6)

# plt.legend(loc = 'upper right')
# Put the legend out of the figure below the x-axis
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), shadow=True, ncol=2)
plt.show()

print(alpha_prior_2, beta_prior_2, alpha_posterior_2, beta_posterior_2)
print(n, y)
```
